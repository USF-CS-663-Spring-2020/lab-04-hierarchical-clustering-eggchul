```python
import pandas as pd
import numpy as np
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,10))
dendrogram(Z)
plt.show()
```
