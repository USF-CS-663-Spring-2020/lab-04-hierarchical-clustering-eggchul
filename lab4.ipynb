```python
import pandas as pd
import numpy as np
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

#### At the distance around 15300, there are two clusters. 

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(30,20))
dendrogram(Z, leaf_font_size= 8)
plt.show()
```

```python
nips_df = pd.read_csv('NIPS_1987-2015.csv')
nips_df
```

```python
nips_df.rename(columns = { "Unnamed: 0" :"type"} , inplace = True)
```

```python
nips_df['Total Count']= df.iloc[:, 1:].sum(axis=1)
nips_df
```

#### In order to We then pick the first 2000 most count words from nips_df 

```python
df2 = nips_df.sort_values(by='Total Count', ascending=False).head(2000).iloc[:,:5812]
```

```python
df2
```

```python
df2 = df2.set_index('type')
del df2.index.name
```

#### At the distance of around 1760, there are three cluster. We could notice that, the word 'algorithm' is isolated from all the other words so far in the top 2000 most count words. The other two main clusters starts at the distance around 1500.

```python
Z = linkage(df2, 'ward')
plt.figure(figsize=(20,250))
#dendrogram(Z)
dendrogram(Z, orientation="left", labels = df2.index, leaf_font_size=8, show_contracted=True)
plt.xlabel('distance')
plt.show()
```
